```python
x = np.array(data[data['sex'] == 'male']['bmi'])
y = np.array(data[data['sex'] == 'female']['bmi'])
t,p_value = stats.ttest_ind(x,y,axis = 0)
             
print(t,p_value)

if p_value < 0.05:
    print('REJECT HO')
else:
    print('accept h0')
             
```

```python
x = np.array(data[data['smoker']== 'yes']['charges'])
y = np.array(data[data['smoker']== 'no']['charges'])
t,p_value = stats.ttest_ind(x,y,axis = 0)
print(t,p_value)
if p_value < 0.05:
    print('REJECT HO')
else:
    print('accept h0')
```
